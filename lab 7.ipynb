```python
import pandas as pd
```

```python
# Weather data for the week
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Temperature (°C)": [30, 28, 32, 29, 35, 33, 31],
    "Humidity (%)": [70, 65, 80, 75, 60, 55, 68],
    "Wind Speed (km/h)": [12, 10, 15, 8, 20, 18, 14],
    "Weather Condition": ["Sunny", "Cloudy", "Rainy", "Cloudy", "Sunny", "Sunny", "Rainy"]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Find the hottest and coldest days of the week
hottest_day = df.loc[df['Temperature (°C)'].idxmax()]
coldest_day = df.loc[df['Temperature (°C)'].idxmin()]
```

```python
df=pd.DataFrame([hottest_day,coldest_day])
```

```python
df
```

```python
high_wind_days = df[df['Wind Speed (km/h)'] > 15]
```

```python
df=pd.DataFrame(high_wind_days)
```

```python
df
```

```python
weather_condition_count = df['Weather Condition'].value_counts()
```

```python
df=pd.DataFrame(weather_condition_count)
```

```python
df
```

```python

average_humidity = df['Humidity (%)'].mean()

print(f"Average Humidity: {average_humidity}")
```
