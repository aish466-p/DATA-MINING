```python
data = pd.DataFrame({
    'Math': [85, 78, 92, 88, 76],
    'Science': [88, 79, 95, 84, 82],
    'English': [75, 83, 89, 92, 80]
})

# correlation matrix
correlation_matrix = data.corr()

# heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['sex'].value_counts().plot(kind='bar')
plt.title('Bar Plot of Sex Distribution')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['age'].hist(bins=20,  alpha=0.7)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sex_counts = df['sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=['Male', 'Female'], autopct='%1.1f%%')
plt.title('Pie Chart of Sex Distribution')
plt.show()
```
